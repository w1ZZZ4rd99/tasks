# Reporting & Visualization (Day 7)

Demonstrates `src/reporting.py`: text / JSON / CSV reports and matplotlib charts (pie, bar, balance-movement). Run from the repository root; artifacts are written to `tmp/`.

```python
from io import BytesIO
from IPython.display import Image

from src import simulation
from src.reporting import ReportBuilder


def show(fig):
    """Render a matplotlib Figure inline (works under the headless Agg backend)."""
    buf = BytesIO()
    fig.savefig(buf, format="png", bbox_inches="tight")
    buf.seek(0)
    return Image(buf.read())


# Reuse the Day 6 simulation to populate a bank with clients, accounts and history.
result = simulation.simulate()
bank, processor = result["bank"], result["processor"]
builder = ReportBuilder(bank, processor)
print(f"{bank.name}: {len(bank.clients)} clients, {len(bank.accounts)} accounts")
```

## Bank report

```python
print(builder.to_text(builder.bank_report()))
```

## Client report (C1)

```python
print(builder.to_text(builder.client_report('C1')))
```

## Risk report

```python
print(builder.to_text(builder.risk_report()))
```

## Charts
### Balance by currency (pie)

```python
show(builder.chart_currency_pie())
```

### Transactions by type (bar)

```python
show(builder.chart_transactions_bar())
```

### Account net flow (line)

```python
show(builder.chart_balance_movement(bank.accounts[0].account_id))
```

## Export to JSON / CSV and save charts (into `tmp/`)

```python
import os

os.makedirs("tmp", exist_ok=True)
print(builder.export_to_json(builder.bank_report(), os.path.join("tmp", "bank.json")))
print(builder.export_to_csv(builder.bank_report()["accounts"], os.path.join("tmp", "accounts.csv")))
for path in builder.save_charts("tmp"):
    print(path)
```
